```python
df = (spark.read
        .option("header",True)
        .csv('/Volumes/world_cup_2026/bronze/raw_files/fifa_world_cup_enhanced_1974_2022.csv'))

display(df)
```

```python
# 1. Aseguramos que exista el esquema (base de datos) 'bronze' dentro de tu catálogo
spark.sql("CREATE SCHEMA IF NOT EXISTS world_cup_2026.bronze")

# 2. Guardamos el DataFrame como una tabla Delta gestionada por Unity Catalog
# Usamos el modo 'overwrite' para inicializarla o 'append' si llegaran más torneos después.
(df.write
 .format("delta")
 .mode("overwrite")
 .option("overwriteSchema", "true")
 .saveAsTable("world_cup_2026.bronze.matches_raw"))

print("¡Tabla Bronze creada exitosamente en Unity Catalog!")
```
