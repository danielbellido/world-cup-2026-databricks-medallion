```python
import kagglehub



# Download latest version

path = kagglehub.dataset_download("samyakrajbayar/fifa-world-cup")



print("Path to dataset files:", path)
```

```python
import os

# Listar los archivos descargados en la ruta local
archivos = os.listdir(path)
print("Archivos dentro de la carpeta:", archivos)
```

```python
import os
import shutil

# 1. Definimos las rutas
ruta_destino_uc = "/Volumes/world_cup_2026/bronze/raw_files/"

# 2. Aseguramos que la carpeta destino en Unity Catalog exista
os.makedirs(ruta_destino_uc, exist_ok=True)

# 3. Copiamos usando la librería nativa de Python
for archivo in os.listdir(path):
    ruta_origen_local = os.path.join(path, archivo)
    ruta_final = os.path.join(ruta_destino_uc, archivo)
    
    # shutil no necesita el prefijo "file:"
    shutil.copy(ruta_origen_local, ruta_final)
    print(f"Copiado con éxito a Unity Catalog: {ruta_final}")
```
