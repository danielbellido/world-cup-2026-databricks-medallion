```python
from pyspark.sql.functions import *
```

```python
spark.sql("CREATE SCHEMA IF NOT EXISTS world_cup_2026.silver")
```

```python
df_bronze = spark.sql("SELECT * FROM world_cup_2026.bronze.matches_raw")

display(df_bronze)
```

### TRANSFORMAR COLUMNAS A SU TIPO CORRECTO

```python
df_silver_matches = (df_bronze
                    .withColumn("date",to_date(col("date"),"yyyy-MM-dd"))
                    .withColumn("year",col("year").cast("integer"))
                    .withColumn("home_goals",col("home_goals").cast("integer"))
                    .withColumn("away_goals",col("away_goals").cast("integer"))
                    .withColumn("goal_difference",col("goal_difference").cast("integer"))
                    .withColumn("total_goals",col("total_goals").cast("integer"))
                    .withColumn("home_xg",col("home_xg").cast("decimal(10,2)"))
                    .withColumn("away_xg",col("away_xg").cast("decimal(10,2)"))
                    .withColumn("possession_home",col("possession_home").cast("integer"))
                    .withColumn("possession_away",col("possession_away").cast("integer"))
                    .withColumn("shots_home",col("shots_home").cast("integer"))
                    .withColumn("shots_away",col("shots_away").cast("integer"))
                    .withColumn("shots_ontarget_home",col("shots_ontarget_home").cast("integer"))
                    .withColumn("shots_ontarget_away",col("shots_ontarget_away").cast("integer"))
                    .withColumn("corners_home",col("corners_home").cast("integer"))
                    .withColumn("corners_away",col("corners_away").cast("integer"))
                    .withColumn("fouls_home",col("fouls_home").cast("integer"))
                    .withColumn("fouls_away",col("fouls_away").cast("integer"))
                    .withColumn("yellow_cards_home",col("yellow_cards_home").cast("integer"))
                    .withColumn("yellow_cards_away",col("yellow_cards_away").cast("integer"))
                    .withColumn("red_cards_home",col("red_cards_home").cast("integer"))
                    .withColumn("red_cards_away",col("red_cards_away").cast("integer"))
                    .withColumn("attendance",col("attendance").cast("integer"))
                    .withColumn("penalty_home", coalesce(col("penalty_home").cast("integer"), lit(0)))
                    .withColumn("penalty_away", coalesce(col("penalty_away").cast("integer"), lit(0)))
                    .withColumn("tournament_format",col("tournament_format").cast("integer")))
    
display(df_silver_matches)
```

```python
# 4. Guardamos en la capa Silver como una tabla optimizada
(df_silver_matches.write
 .format("delta")
 .mode("overwrite")
 .saveAsTable("world_cup_2026.silver.fact_matches"))
```
